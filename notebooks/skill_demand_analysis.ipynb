```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import re

# Step 2: Load Dataset
df = pd.read_csv('C:/Users/jagadeeswari/OneDrive/Desktop/Job-Market-Skill-Demand-Analysis/data/data job posts.csv/data job posts.csv')

print("Dataset Preview:")
display(df.head())
print(df.columns)

# Step 3: Combine all job descriptions
df.columns = df.columns.str.strip()

all_descriptions = " ".join(
    df["JobDescription"].fillna("").astype(str)
)


# Step 4: Predefined Skill List
skills_list = [
    "SQL", "Python", "Power BI", "Excel", "Tableau",
    "Machine Learning", "Deep Learning",
    "TensorFlow", "AWS", "DAX",
    "Data Visualization"
]

# Step 5: Count Skill Frequency
skill_count = {}

for skill in skills_list:
    skill_count[skill] = len(re.findall(skill, all_descriptions, re.IGNORECASE))

# Convert to DataFrame
skills_df = pd.DataFrame(skill_count.items(), columns=["Skill", "Count"])
skills_df = skills_df.sort_values(by="Count", ascending=False)

print("\nTop Skills:")
display(skills_df)

# Step 6: Visualization
plt.figure()
plt.bar(skills_df["Skill"], skills_df["Count"])
plt.xticks(rotation=45)
plt.title("Top In-Demand Skills")
plt.xlabel("Skills")
plt.ylabel("Frequency")
plt.tight_layout()
plt.savefig("../outputs/top_skills_chart.png")
plt.show() 
```
